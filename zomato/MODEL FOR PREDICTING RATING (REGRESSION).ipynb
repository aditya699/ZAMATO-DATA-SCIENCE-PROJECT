```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('C:/Users/aditya/Desktop/zomato.csv')
data.head()
```

```python
data.drop(['url'],axis=1,inplace=True)
```

```python
data.drop(['address'],axis=1,inplace=True)
```

```python
data.drop(['phone'],axis=1,inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
#REMOVING NAN VALUES
data.isnull().sum()
data.dropna(how='any',inplace=True)
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
data.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type','listed_in(city':'city'},inplace=True)
```

```python
data['cost']=data['cost'].astype(str)
data['cost']=data['cost'].apply(lambda x:x.replace(',',''))
data['cost']=data['cost'].astype(float)
```

```python
data['rate'].unique()
```

```python
data['rate']=data['rate'].replace('NEW',np.nan)
data['rate'].dropna(inplace=True)
```

```python
remove_slash=lambda x:x.replace('/5','') if type(x)==np.str else x
data.rate=data.rate.apply(remove_slash).str.strip().astype('float')
data['rate']
```

```python
data['cost']=data['cost'].replace('.','')
```

```python
data['cost'].unique()
```

```python
data.drop(['name','dish_liked','cuisines','reviews_list','menu_item','listed_in(city)'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data['rate'].unique()
```

```python
data['rate'].fillna(2.55,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
data.head()
```

```python
y=data['rate']
x=data.drop('rate',axis=1)
```

```python
x.head()
```

```python
y.head()
```

# MODEL BULIDING

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.1,random_state=353)
x_train.head()
```

```python
lr = LinearRegression()
```

```python
model=lr.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
r2
```

```python
#SO WE GET A BASELINE ACCURACY OF 34%
```

```python
from sklearn.tree import DecisionTreeRegressor 
dt=DecisionTreeRegressor()
model_dt=dt.fit(x_train,y_train)
y_pred=model_dt.predict(x_test)
r2 = r2_score(y_test, y_pred)
r2
```

```python
# SO WITH DT REGRSSOR WE GET A ACCURACY OF 94%
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rfr=RandomForestRegressor(n_estimators=1000)
model_rfr=model_rfr.fit(x_train,y_train)
y_pred=model_rfr.predict(x_test)
r2 = r2_score(y_test, y_pred)
r2
```

```python
#SO WE GET A DECENT ACCRACY.
```
